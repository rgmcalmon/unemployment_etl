```python
# Dependencies
import pandas as pd
import os
import csv
# Import SQLAlchemy `automap` and other dependencies
from sqlalchemy import create_engine, inspect
```

```python
unemployment_by_county_df = pd.read_csv('raw_data/unemployment_by_county.csv.gz')
crime_data_with_population_df = pd.read_csv('raw_data/crime_data_w_population_and_crime_rate.csv')
states_df = pd.read_csv('raw_data/states.csv')
```

```python
unemployment_by_county_df.dropna(inplace = True)
unemployment_by_county_df.head()
```

```python
county_state_pair = crime_data_with_population_df['county_name'].str.split(',')
```

```python
def clean(df):
    return df\
    .assign(county_state_pair=df['county_name'].str.split(','))\
    .assign(county=lambda df0: df0['county_state_pair'].apply(lambda l: l[0]))\
    .assign(state=lambda df0: df0['county_state_pair'].apply(lambda l: l[1]))
```

```python
crime_data_with_population_df = clean(crime_data_with_population_df)
```
